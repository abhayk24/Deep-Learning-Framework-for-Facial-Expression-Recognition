```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
results = {
    "Model": [
        "CNN",
        "ResNet50",
        "Vision Transformer"
    ],

    "Accuracy": [
        0.5361,
        0.6199,
        0.6199
    ],

    "Precision": [
        0.5459,
        0.6139,
        0.6123
    ],

    "Recall": [
        0.5361,
        0.6199,
        0.6199
    ],

    "F1 Score": [
        0.5391,
        0.6133,
        0.6117
    ]
}

df = pd.DataFrame(results)

df
```

```python
print("="*70)
print("FACIAL EXPRESSION RECOGNITION MODEL COMPARISON")
print("="*70)

display(df)
```

```python
plt.figure(figsize=(8,5))

bars = plt.bar(
    df["Model"],
    df["Accuracy"]
)

plt.title("Accuracy Comparison")
plt.ylabel("Accuracy")

for bar in bars:
    y = bar.get_height()
    plt.text(
        bar.get_x()+bar.get_width()/2,
        y+0.005,
        f"{y:.4f}",
        ha='center'
    )

plt.ylim(0.45,0.70)

plt.show()
```

```python
plt.figure(figsize=(8,5))

bars = plt.bar(
    df["Model"],
    df["Precision"]
)

plt.title("Precision Comparison")
plt.ylabel("Precision")

for bar in bars:
    y = bar.get_height()
    plt.text(
        bar.get_x()+bar.get_width()/2,
        y+0.005,
        f"{y:.4f}",
        ha='center'
    )

plt.ylim(0.50,0.70)

plt.show()
```

```python
plt.figure(figsize=(8,5))

bars = plt.bar(
    df["Model"],
    df["Recall"]
)

plt.title("Recall Comparison")
plt.ylabel("Recall")

for bar in bars:
    y = bar.get_height()
    plt.text(
        bar.get_x()+bar.get_width()/2,
        y+0.005,
        f"{y:.4f}",
        ha='center'
    )

plt.ylim(0.50,0.70)

plt.show()
```

```python
plt.figure(figsize=(8,5))

bars = plt.bar(
    df["Model"],
    df["F1 Score"]
)

plt.title("F1 Score Comparison")
plt.ylabel("F1 Score")

for bar in bars:
    y = bar.get_height()
    plt.text(
        bar.get_x()+bar.get_width()/2,
        y+0.005,
        f"{y:.4f}",
        ha='center'
    )

plt.ylim(0.50,0.70)

plt.show()
```

```python
x = np.arange(len(df))

width = 0.2

plt.figure(figsize=(10,6))

plt.bar(x-1.5*width,df["Accuracy"],width,label="Accuracy")
plt.bar(x-0.5*width,df["Precision"],width,label="Precision")
plt.bar(x+0.5*width,df["Recall"],width,label="Recall")
plt.bar(x+1.5*width,df["F1 Score"],width,label="F1 Score")

plt.xticks(x,df["Model"])

plt.ylim(0.45,0.70)

plt.legend()

plt.title("Performance Comparison of CNN, ResNet50 and Vision Transformer")

plt.show()
```
